```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
%matplotlib inline
# Simulate EEG data
np.random.seed(42)
t = np.arange(0, 10, 0.01)
eeg_data = 5 * np.sin(2 * np.pi * 1 * t) + 2 * np.sin(2 * np.pi * 5 * t) + np.random.normal(scale=2, size=len(t))

# Plot the simulated EEG data
plt.figure(figsize=(10, 4))
plt.plot(t, eeg_data, label='Simulated EEG Data')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.title('Simulated EEG Data')
plt.show()

# Apply Moving Average (MA)
ma_order = 3
ma_model = sm.tsa.SARIMAX(eeg_data, order=(2, 0, ma_order), enforce_stationarity=False)
ma_results = ma_model.fit()

# Apply Auto-Regressive (AR)
ar_order = 2
ar_model = sm.tsa.SARIMAX(eeg_data, order=(ar_order, 0, 2), enforce_stationarity=False)
ar_results = ar_model.fit()

# Apply Auto-Regressive Moving Average (ARMA)
arma_order = (2,0, 3)  # (AR order, difference,MA order)
arma_model = sm.tsa.SARIMAX(eeg_data, order=arma_order, enforce_stationarity=False)
arma_results = arma_model.fit()

# Plot the original EEG data and fitted models
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(t, eeg_data, label='Original EEG Data')
plt.title('Original EEG Data')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(t, ma_results.fittedvalues, label=f'MA({ma_order}) Fitted Values')
plt.title(f'MA({ma_order}) Model')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(t, ar_results.fittedvalues, label=f'AR({ar_order}) Fitted Values')
plt.title(f'AR({ar_order}) Model')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(t, arma_results.fittedvalues, label=f'ARMA({arma_order}) Fitted Values')
plt.title(f'ARMA({arma_order}) Model')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Simulate EEG data
np.random.seed(42)
t = np.arange(0, 10, 0.01)
eeg_data = 5 * np.sin(2 * np.pi * 1 * t) + 2 * np.sin(2 * np.pi * 5 * t) + np.random.normal(scale=2, size=len(t))

# Plot the simulated EEG data
plt.figure(figsize=(10, 4))
plt.plot(t, eeg_data, label='Simulated EEG Data')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.title('Simulated EEG Data')
plt.show()

# Apply Auto-Regressive Moving Average (ARMA)
ar_order = 2
ma_order = 3
arma_order = (ar_order, 1, ma_order)  # (AR order, difference, MA order)
arma_model = sm.tsa.SARIMAX(eeg_data, order=arma_order, enforce_stationarity=False)
arma_results = arma_model.fit()

# Plot the original EEG data and fitted model
plt.figure(figsize=(10, 6))
plt.plot(t, eeg_data, label='Original EEG Data')
plt.plot(t, arma_results.fittedvalues, label=f'ARMA({ar_order},{ma_order}) Fitted Values')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.title(f'ARMA({ar_order},{ma_order}) Model Fit to EEG Data')
plt.show()

# Display model summary
print(arma_results.summary())
```
