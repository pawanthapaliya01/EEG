```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, FastICA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn import datasets

# Generate synthetic EEG data for demonstration purposes
# Replace this with your actual EEG data
n_samples, n_features = 100, 20
eeg_data, labels = datasets.make_classification(n_samples=n_samples, n_features=n_features, n_classes=2, random_state=42)

# Principal Component Analysis (PCA)
pca = PCA(n_components=2)
eeg_pca = pca.fit_transform(eeg_data)

# Linear Discriminant Analysis (LDA)
# Ensure that n_components is less than or equal to min(n_features, n_classes - 1)
n_components_lda = min(n_features, len(np.unique(labels)) - 1)
lda = LinearDiscriminantAnalysis(n_components=n_components_lda)
eeg_lda = lda.fit_transform(eeg_data,labels)

# Independent Component Analysis (ICA)
ica = FastICA(n_components=2)
eeg_ica = ica.fit_transform(eeg_data)

# Plotting the results
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.scatter(eeg_pca[:, 0], eeg_pca[:, 1], c=labels, cmap='viridis')
plt.title('Principal Component Analysis (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.subplot(132)
plt.scatter(eeg_lda[:, 0], eeg_lda[:, 0], c=labels, cmap='viridis')
plt.title('Linear Discriminant Analysis (LDA)')
plt.xlabel('Linear Discriminant 1')
plt.ylabel('Linear Discriminant 2')

plt.subplot(133)
plt.scatter(eeg_ica[:, 0], eeg_ica[:, 1], c=labels, cmap='viridis')
plt.title('Independent Component Analysis (ICA)')
plt.xlabel('Independent Component 1')
plt.ylabel('Independent Component 2')

plt.tight_layout()
plt.show()
```
